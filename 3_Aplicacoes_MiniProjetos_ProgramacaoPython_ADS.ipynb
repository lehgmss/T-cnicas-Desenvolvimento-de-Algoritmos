# **Seção 2: E-commerce**

**Catálogo de Produtos**

```python
catalogo = [
    ("Smartphone", 1500, "eletronicos"),
    ("Tênis", 300, "moda"),
    ("Notebook", 4500, "eletronicos"),
    ("Livro", 50, "educacao"),
    ("Fone de Ouvido", 200, "eletronicos")
]

eletronicos = [produto for produto in catalogo if produto[2] == "eletronicos"]
assert len(eletronicos) == 3
print("Produtos eletrônicos:", eletronicos)
```

**Carrinho de Compras**

```python
carrinho = {}
carrinho["Notebook"] = 1
carrinho["Fone de Ouvido"] = 2
del carrinho["Notebook"]

assert "Notebook" not in carrinho
assert carrinho["Fone de Ouvido"] == 2
print("Carrinho:", carrinho)
```

**Cálculo de Total com desconto**

```python
precos = [1500, 300, 4500]
precos_com_desconto = [p * 0.9 for p in precos]
total = sum(precos_com_desconto)

assert round(total, 2) == round(1500*0.9 + 300*0.9 + 4500*0.9, 2)
print(f"Total com desconto: R$ {total:.2f}")
```

**Estoque por categoria**

```python
estoque = {
    "Eletrônicos": ["Smartphone", "Notebook", "Fone de ouvido"],
    "Moda": ["Tênis", "Camisa"],
    "Educação": ["Livro"]
}

categoria_maior = max(estoque, key=lambda c: len(estoque[c]))
assert categoria_maior == "Eletrônicos"
print("Categoria com mais produtos:", categoria_maior)
```

# **Seção 3: Jogos**

**Ranking de Jogadores**

```python
ranking = [("Leticia", 800), ("Bianca", 500), ("Giselle", 550)]
ranking.sort(key=lambda x: x[1], reverse=True)

assert ranking[0][0] == "Leticia"
print("Top 3 jogadores:", ranking[:3])
```

**Inventário de Itens**

```python
inventario = {
    "Revolver": (1, "raro"),
    "Balas": (25, "comum"),
    "Colete": (2, "épico")
}

raros = [item for item, (_, raridade) in inventario.items() if raridade in ["raro", "épico"]]
assert "Revolver" in raros and "Colete" in raros
print("Itens raros:", raros)
```

**Histórico de Partidas**

```python
partidas = [
    ("Leticia", "Bianca", "Giselle"),
    ("Leticia", "Daniele", "Leticia"),
    ("Bianca", "Giselle", "Daniele")
]

vitorias = {}

for partida in partidas:
    for jogador in partida:
        vitorias[jogador] = vitorias.get(jogador, 0) + 1

assert vitorias["Leticia"] == 3
print("Vitórias:", vitorias)
```

**Configurações do Jogo**

```python
config = {
    "volume": 70,
    "dificuldade": "hard",
    "idioma": "pt-br"
}

config["volume"] = 50
config["idioma"] = "en-us"


assert config["volume"] == 50
print("Configurações:", config)
```

# **Seção 4: Playlists**

**Playlists de Músicas**

```python
playlist = [
    {"nome": "Baby", "artista": "Justin Bieber", "duracao": 9.3},
    {"nome": "Positions", "artista": "Ariana Grande", "duracao": 4.5},
    {"nome": "Purpose", "artista": "Justin Bieber", "duracao": 5.5}
]

longas = [m["nome"] for m in playlist if m["duracao"] > 5]
assert "Purpose" in longas and "Baby" in longas
print("Músicas com mais de 5 minutos:", longas)
```

**Favoritar Músicas**

```python
favoritas = {
    "Baby": True,
    "Positions": False,
    "Purpose": True
}

favoritas_listadas = [nome for nome, fav in favoritas.items() if fav]
assert len(favoritas_listadas) == 2
print("Favoritas:", favoritas_listadas)
```

**Estatísticas da Playlist**

```python
musicas = [("Baby", 9.1), ("Positions", 4.5), ("Purpose", 5.5)]
total = sum([dur for _, dur in musicas])
media = total / len(musicas)

assert round(media, 2) == round(total / 3, 2)
print(f"Total: {total:.2f} min | Média: {media:.2f} min")
```

**Organizar por Artista**

```python
por_artista = {
    "Justin Bieber": ["Baby", "Purpose"],
    "Ariana Grande": ["Positions"]
}

artista_top = max(por_artista, key=lambda a: len(por_artista[a]))
assert artista_top == "Justin Bieber"
print("Artista com mais músicas:", artista_top)
```
